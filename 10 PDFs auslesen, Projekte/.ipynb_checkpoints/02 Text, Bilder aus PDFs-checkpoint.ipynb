## Alle Files auslesen

```python
lst = []#leere liste
mistakes = []#fehler
for pdf in tqdm(pdfs):
    try:
        text = textract.process("pdfs/"+pdf, method='pdfminer', encoding='ISO-8859-1')#text herausziehen plus encoding(brauchts nicht zwingend, also die beiden letzten)
        lst.append(text)#zieht text heraus
    except:
        mistakes.append(pdf)
```

## Alles zusammen

```python
lst = []
#Rohe Analyse selbe struktur wie oben einfach erweitert um eintrag, dass wenn textlänge grösser als 40 handelt es sich um ein Bild
for pdf in tqdm(pdfs):
    try:
        text = textract.process("pdfs/"+pdf, method='pdfminer', encoding='ISO-8859-1')
        text = text.decode("ISO-8859-1").replace("\n", " ")#decode braucht es, replace ist rein optisch
        if len(text) < 40:
            text = textract.process("pdfs/"+pdf, method='tesseract', encoding='ISO-8859-1')
            text = text.decode("ISO-8859-1").replace("\n", " ")#decode braucht es, replace ist rein optisch
            
        mini_dict = {'Text':text,
                     'File':pdf}
        lst.append(mini_dict)
    except:
        mini_dict = {'Text':'Fehlermeldung',
                     'File':pdf}
        lst.append(mini_dict)
```

```python
pd.DataFrame(lst)
```

```python
df = pd.DataFrame(lst)
```

```python
# damit kann man nun basteln, bsp. textlänge rausziehen
def length(elem):
    elem = len(elem)
    return elem
df ["Text Länge"] = df["Text"].apply(length)
```

```python
df
```

```python
#Erste Säuberung
df = pd.DataFrame(lst)
def date(elem):
    elem = elem[3:11]
    return elem
df['date'] = df['File'].apply(date)
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df.index = df['date']
#Suchen wir noch nach Implantaten
df['implant'] = df['Text'].str.contains('implant')
df.to_csv('datafile.csv')
```
