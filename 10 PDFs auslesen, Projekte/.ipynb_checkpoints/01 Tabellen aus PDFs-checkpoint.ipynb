### Und nun hängen wir alls zusammen

```python
#eine leere df bauen
ff = pd.DataFrame()#leeres dataframe kreieren
headers = ['Gemeinde', 'Sachschaden', 'Personenschaden', 'Getötete', 
           'Schwerverletzte', 'Lebensbedrohlich Verletzte', 
           'Erheblich Verletzte', 'Leichtverletzte']

#Und nun durch alles iterieren
#https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
for number in range(0,8):
    temp = tables_stream[number].df #alle seiten
    temp = temp[9:]#schau die oberen 8 nicht an, da dort müll drin ist
    temp.columns = headers#kolumnen definieren
    frames = [ff, temp]#die beiden frames auswählen ff ist die leere liste, temp ist ab zeile langenthal
    ff = pd.concat(frames)
```

```python
ff.info()
#oder auch ff.describe()
#oder auch ff.dtypes
```

```python
## Mit leeren Strings umgehen. wenn "" gib 0 zurück, 
#wenn nicht gib die entsprechende zahl in der tabelle (hier definiert als elem) zurück
def clean(elem):
    if elem == "":
        return 0
    else:
        return elem
ff = ff.applymap(clean)
```

```python
## Die Gemeinden zwischenzeitlich löschen und dann wieder reinnehmen bei der Umwandlung in Zahlen, 
#ff ist das gesamte dataframe(gesamte datebank)
gem = ff['Gemeinde'] #zuerst die spalte abspeichern
del ff['Gemeinde'] #spalte wird gelöscht
ff = ff.astype(int) #dann string umwandeln in int (zahl)
ff['Gemeinde'] = gem #wieder anhängen
ff = ff[['Gemeinde', 'Sachschaden', 'Personenschaden', 'Getötete', 
           'Schwerverletzte', 'Lebensbedrohlich Verletzte', 
           'Erheblich Verletzte', 'Leichtverletzte']].copy() ##Reihenfolge, nur kosmetisch
```

```python
#index anpassen
ff = ff.reset_index()
del ff['index']
```
