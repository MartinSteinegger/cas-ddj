Arbeiten Sie in Gruppen, um ein Ergebnis zu erarbeiten!

```python
anfangswert = df.iloc[0]
anfangswert
```

```python
df_neu = ((df / anfangswert) - 1) * 100
```

```python
df_neu
```

```python
endwert = df_neu.iloc[-1].sort_values(ascending=False)
endwert
```

```python
df_neu.iloc[-1].sort_values(ascending=False).plot(kind='bar', title='Kursveränderung in Prozent')
```

```python
endwert = (df_neu.iloc[-1] * 5000 / 100).sort_values(ascending=False)
```
